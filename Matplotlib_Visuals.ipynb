```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
plt.grid()
plt.xlabel('X-axis: Time')
plt.ylabel('Y-axis: Amplitude')
plt.title('Basic line plot of a Sine Wave')
plt.figure(figsize = (8,6))
plt.show()
```

```python

x = np.linspace(0,10,100)
y = np.sin(x)
plt.grid()
plt.plot(x,y)
plt.figure(figsize=(5,5))

```

```python
x = np.random.rand(1000) *100
y = np.random.rand(1000) *10
plt.scatter(x,y, marker = 'o')
```

```python
data =  np.random.rand(100)
plt.figure(figsize=(6,4))
plt.hist(data, bins = 100, alpha=0.98)  # 'bins' sets the number of bins, 'alpha' sets transparency
plt.show()
print(type(data))
```

```python
data1 = np.random.randn(1000)
data2 = np.random.randn(1000)+2
plt.hist([data1,data2],bins=30,alpha = 0.7,
         label=['data1', 'data2'], 
         color = ['blue','green'])
plt.xlabel('value')
plt.ylabel('Frequency')
plt.title('Histogram of Two Datasets')
plt.legend()
plt.show()
```

```python
size = [13.56,65,34,78]
explode = [0.1,0,0,0.05]
plt.figure(figsize=(9,9))
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
plt.pie(size,explode=explode,labels=labels,autopct = '%1.1f%%',startangle = 90,colors = color)
plt.legend(labels,title ='fruits',loc='upper right')
plt.show()
```
